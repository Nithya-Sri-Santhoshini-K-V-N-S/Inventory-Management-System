**Read the Inventory**

```python
import json
fd=open("record.json","r")
r=fd.read()
fd.close()
records=json.loads(r)
records
```

**Purchase**

```python
print("----------------------------------------------")
print("----------------------------------------------")
print("           SUN GROCERY STORE                  ")
print("      45,Beach side road, Chennai-02          ")
print("         Contact: 76555543244                 ")
print("----------------------------------------------")
print("----------------------------------------------")

customer_name=str(input("Name:"))

ui_prod_id=str(input("Enter Product ID:"))
ui_quant=int(input("Enter Quantity:"))

if ui_quant > records[ui_prod_id]["Quantity"]:
  print("Stock not available.Only",records[ui_prod_id]["Quantity"],"nos available in the store")
else:
  print("Product Name:", records[ui_prod_id]["Product"])
  print("Rate:", records[ui_prod_id]["Rate"])
  print("Manufacturing Date:", records[ui_prod_id]["Manufacturing Date"])
  print("Expiry Date:", records[ui_prod_id]["Expiry Date"])
  print("----------------------------------------------")
  print("Billing Amount:", records[ui_prod_id]["Rate"] * int(ui_quant))
  print("----------------------------------------------")
  records[ui_prod_id]["Quantity"]=records[ui_prod_id]["Quantity"]-ui_quant

import time
asq=time.ctime()
print(asq)
```

**Update Inventory after purchase**

```python
js=json.dumps(records)
fd=open("record.json","w")
fd.write(js)
fd.close()
```

**Sales JSON**

```python
sqn={"Customer Name":customer_name,"Product ID":ui_prod_id,"Product Name":records[ui_prod_id]["Product"],"Quantity":ui_quant, "Billing Amount":records[ui_prod_id]["Rate"] * int(ui_quant),"Transaction Time":asq}
```

```python
sales={}
sales[str(len(sales)+1)] = sqn
sale=json.dumps(sales)
fd=open("sales.json","w")
fd.write(sale)
fd.close()
```
